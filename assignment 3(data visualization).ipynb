```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("company_sales_data.csv")
 
plt.plot(data['month_number'])
plt.plot(data['total_profit'])

plt.title("line Plot")
 
plt.xlabel('month')
plt.ylabel('profit')
 
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("company_sales_data.csv")
plt.plot(data['month_number'],data['facewash'])
plt.plot(data['month_number'],data['facecream'])
plt.plot(data['month_number'],data['toothpaste'])
plt.plot(data['month_number'],data['bathingsoap'])
plt.plot(data['month_number'],data['shampoo'])
plt.plot(data['month_number'],data['moisturizer'])
#plt.plot(data['month_number'])

plt.title("line Plot")
 
plt.xlabel('month')
plt.ylabel('profit')
 
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data=pd.read_csv("company_sales_data.csv")
plt.scatter(data['facecream'] ,data['total_profit'],c='green')
plt.scatter(data['facewash'] ,data['total_profit'],c='red')
plt.scatter(data['toothpaste'] ,data['total_profit'],c='blue')
plt.scatter(data['shampoo'] ,data['total_profit'],c='yellow')
plt.title("Scatter Plot") 
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv("company_sales_data.csv")
total_sales = data.drop(['month_number', 'total_units', 'total_profit'], axis=1).sum()
plt.bar(total_sales.index, total_sales.values)
plt.title('Total Sales for Each Product')
plt.xlabel('Product')
plt.ylabel('Total Sales')
plt.show()
```
